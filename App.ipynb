# Outputs

### Basic Preprocessing

```
df = pd.DataFrame({'Nitrogen' : [Nitrogen], 'phosphorus' : [Phosphorus], 'potassium' : [Potassium], 'temperature' : [float(Temperature_in_c)], 'humidity' : [float(Humidity)], 'ph' : [float(pH)], 'rainfall' : float(Rainfall_in_mm)})
z = model.predict(df)
```

```
from sklearn.ensemble import IsolationForest
import pandas as pd

df2 = pd.read_csv('data/df2.csv')
df2.drop(columns = ['Unnamed: 0'],inplace=True)
df2 = df2[df2['CropHealth'] == int(z[0])]
df2.drop(columns = ['CropHealth'],inplace=True)

iso_forest = IsolationForest(contamination=0.1, random_state=42)
iso_forest.fit(df2)

anomaly_score = iso_forest.decision_function(df)

anomaly_threshold = -0.06
anomaly_threshold2 = -0.04
```

### Condition and Anomaly Identification

```
if anomaly_score[0] < anomaly_threshold and z[0] == 1:

    print("It is in good condition. However it is an Anomaly.")

elif anomaly_score[0] < anomaly_threshold and z[0] == 0:

    print("It is in bad condition. However it is an Anomaly.")

elif anomaly_score[0] < anomaly_threshold2 and z[0] == 1:

    print('It is in good condition. However it is an possible early warning that its health is deteriorating.')

elif anomaly_score[0] < anomaly_threshold2 and z[0] == 0:

    print('It is in bad condition. However it is an possible early detection that its health is improving.')

elif z[0] == 1:

    print("Good Condition.")

else:

    print("Bad Condition.")
```

### General Insights

```
with open('data/avg.txt','r') as f:

    avgstr = f.read().split(', ')

avg = list(map(float,avgstr))
avg = [round(i,2) for i in avg]
```

```
lim = [6,15,12,5,13,1,30]
```

```
res = [(avg[i] - df.loc[0].to_list()[i]) for i in range(7)]
```

```
otl = [abs(res[i]) > lim[i] for i in range(7)]
d = 'Nitrogen Phosphorus Potassium Temperature Humidity pH Rainfall'.split()
gm = sum([1 for i in otl if i])
```

```
for x,i in  enumerate(otl):

    if i:

        print(f'{d[x]} is {round(abs(res[x]))} units {"higher" if int(res[x]) < 0 else "lower"} than average of {round(avg[x],1)}')

if gm == 0:

    print('No visible insights to uncover.')
```

### Fertilizers Info

```
NumFertilizer , CropPeriod = tuple(sqltable[sqltable['Crop'] == Crop][['NumFertilizer','CropPeriod']].iloc[0].to_list())
NumFertilizer += 1
```

```
SownDate = SownDated.date()
CurrentDate = datetime.date.today()
CurrentCropPeriod = (CurrentDate - SownDate).days
```

```
FertilizerFreq = (CropPeriod * 30) / NumFertilizer
TimeLeftToFertilizerInDays = int(CurrentCropPeriod % FertilizerFreq)
```

```
CurrFertilizer = NumFertilizer - NumFertilizerUsed
```

```
if CurrFertilizer > 0:
    
    und = []
    hig = []

    for x,i in  enumerate(otl[:3]):

        if i and res[x] > 0:

            und.append(d[x])

        elif i and res[x] < 0:

            hig.append(d[x])
            
    if len(und) > 0:
        
        print(f"Increase the amount of {', '.join(und)} fertilizer{'s' if len(und) > 1 else ''}", end = '')

    if (len(und) > 0) and (len(hig) > 0):

        print(' and ', end = '')

    if len(hig) > 0:
        
        print(f"Decrease the amount of {', '.join(hig)} fertilizer{'s' if len(hig) > 1 else ''}", end = '')

    ndays = 'today' if TimeLeftToFertilizerInDays == 0 else 'tomorrow' if TimeLeftToFertilizerInDays == 1 else f'in {str(TimeLeftToFertilizerInDays)} days'

    if (len(und) > 0) or (len(hig) > 0):

        print(f' when using fertilizers {ndays}.')

    else:

        print(f'Use usual load of fertilizers when using fertilizers {ndays}.')
```

### Harvest Info

```
CropPeriodLeft = (CropPeriod * 30) - CurrentCropPeriod
HarvestDate = SownDate + datetime.timedelta(days=(CropPeriod * 30))
print('Harvest Date :', HarvestDate.strftime("%m/%d/%Y"))
print(f'Number of Application of fertilizers left : {NumFertilizer - NumFertilizerUsed - 1}')
print(f'Time left until harvest t-minus {CropPeriodLeft} days.')
```

### Rainfall Info

```
if otl[6]:

    rf = int(res[6])

    if rf < 0:

        print(f'High Rainfall of {-rf} cm higher than average of {int(avg[6])} cm. Check for water logging.')

    elif rf > 0:

        print('Low Rainfall of {rf} cm lower than average of {int(avg[6])} cm. Irrigate the crops soon.')

else:

        print('Sufficient Rainfall.')
```

### Pest Management

```
def conditions(res,otl):

    a = 0
    a += res[0] * 3 if res[0] > 0 else 0
    a += res[1] * 3 if res[1] > 0 else 0
    a += res[2] * 3 if res[2] > 0 else 0
    a += 100 if otl[3] else 0
    a += 100 if res[4] < 0 else 0
    a += 100 if res[6] < 0 else 0

    return a

if conditions(res,otl) > 250:

    print('Possible Pest Infestation.\n(Low Nutrients, Extreme Temperatures or High Rainfall \\ Humidity)')

else:

    print('Very low probability of pest infestation.')
```
